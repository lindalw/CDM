```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
```

```python
def img_selector(target):
    target = str(target)
    
    if len(target) == 5:
        img = 'COCO_train2014_0000000' + target + '.jpg'
    else:
        img = 'COCO_train2014_00000000' + target + '.jpg'
    
    image_folders = next(os.walk('.'))[1]

#   loop throw all folders
    for folder in image_folders:
#         print("FOLDER", folder)
#         print("####")

#       loop throw the documents in folders
        for documents in os.walk(folder):
#             print("documents", documents[2])

#           loop throw images in folder
            for doc in documents[2]:

                if doc == img:
                    img = folder + "/" + img
                    image = mpimg.imread(img)
                    plt.imshow(image)
                    plt.show()
                    print("DOMAIN:", folder)
    
    
```

```python
imgs = ['468357', '93469',
  '341060',
  '82894',
  '483794',
  '259745',
  '14238',
  '522129',
  '96757',
  '380128']
for i in imgs:
    img_selector(i)
```
