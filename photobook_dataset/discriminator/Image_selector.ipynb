```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
```

```python
def img_selector(target):
    target = str(target)

    
    if len(target) == 5:
        img = 'COCO_train2014_0000000' + target + '.jpg'
    else:
        img = 'COCO_train2014_000000' + target + '.jpg'
        
    print(img)
    
    image_folders = next(os.walk('../images/images'))[1]


#       loop throw the documents in folders
    for documents in os.walk('../images/images'):
        try:
            current_folder = str(documents[0]).split("\\")[1]
#             print(current_folder)
        except:
            try:
                current_folder = str(documents[0]).split('/')[3]
            except:
                None
                        
        print(str(documents[0]).split('/'))
#       loop throw images in folder
        for doc in documents[2]:
#             print("doc", doc)

            if doc == img:
#                 print("FOUND")
                img = '../images/images/' + current_folder + "/" + img
                image = mpimg.imread(img)
                plt.imshow(image)
                plt.show()
                print("DOMAIN:", current_folder)
    
    
```

```python
imgs = ['468357',
        '93469',
  '341060',
  '82894',
  '483794',
  '259745',
  '14238',
  '522129',
  '96757',
  '380128']
for i in imgs:
    img_selector(i)
```
